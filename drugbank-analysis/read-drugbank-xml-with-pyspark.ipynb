```python
import pandas as pd
from pyspark.sql import SparkSession
import os
```

```python
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages com.databricks:spark-xml_2.12:0.12.0 pyspark-shell'
```

```python
my_spark_session = (
    SparkSession
    .builder
    .appName("XML_Import_Test")
    .master("local[*]")
    .getOrCreate()
)
```

```python
df2 = (
    my_spark_session
        .read
        .format('xml')
        .option('rootTag', 'drugbank')
        .option('rowTag', 'drug')
        .load('data/drugbank/drugbank-database-jan2021.xml')
)
```

```python
df2.printSchema()
```

```python
my_spark_session.stop()
```
