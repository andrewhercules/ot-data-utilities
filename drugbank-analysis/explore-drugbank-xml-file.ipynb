```python
import xml.etree.ElementTree as ET

ns = '{http://www.drugbank.ca}'

all_drugs_data = []

# get an iterable
context = ET.iterparse("database.xml", events=("start", "end"))

is_first = True

for event, elem in context:
    # get the root element
    if is_first:
        root = elem
        is_first = False
    if event == "end":
#         print('*** New Entry ***')
#         print(elem.findtext(ns + "name"))
        drug_name = elem.findtext(ns + "name")
        all_drugs_data.append(drug_name)
    root.clear()
```

```python
# import library to parse XML
import xml.etree.ElementTree as ET

# parse full database and load into memory
tree = ET.parse('data/database.xml')
root = tree.getroot()

# access record for Insulin human (https://go.drugbank.com/drugs/DB00030)
item = tree.getroot()[28]

# set namespace
ns = '{http://www.drugbank.ca}'

# find external identifiers 
db_ext_ids = item.findall('{ns}external-identifiers/{ns}external-identifier'.format(ns = ns))

# iterate through list of external identifiers entry and print object with resource and identifier
for ext_id in db_ext_ids:
    external_id_obj = {}
    external_id_obj['resource'] = ext_id.findtext(ns + 'resource')
    external_id_obj['url'] = ext_id.findtext(ns + 'identifier')
    print(external_id_obj)
```
